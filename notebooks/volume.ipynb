```python
import numpy as np
import matplotlib.pyplot as plt

from looper.runner.dsp import SignalProcessor
from looper.runner.config import Config
```

```python
config = Config()
dsp = SignalProcessor(config)
```

```python
sine = dsp.sine()
```

```python
plt.plot(sine)
```

```python
dsp.compute_chunk_loudness(sine)  # sine wave should have 0 dBFS
```

```python
amplified = dsp.amplify(sine, -60)
plt.plot(amplified)
```

```python
dsp.compute_chunk_loudness(amplified)  # measured loudness is same as applied amplification
```

```python
sines = [sine] * 10
```

```python
dsp.compute_loudness(sines)  # loudness of chunks is the same as the signle chunk
```
